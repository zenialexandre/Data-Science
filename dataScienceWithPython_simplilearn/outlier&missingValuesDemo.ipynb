```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dataset = load_diabetes()
print(dataset)

dataset.data
dataset.target

dataset['feature_names']

df = pd.DataFrame(data=np.c_[dataset['data'], dataset['target']], columns = dataset['feature_names'] + ['target'])
print(df)

df.isnull().any() # verifying if we have any missing data

# getting the outliers elements
for column in df:
  plt.figure()
  df.boxplot([column])
```
