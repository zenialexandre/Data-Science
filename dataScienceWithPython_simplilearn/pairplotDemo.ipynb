```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# import the csv file with data
url = 'SampleCSVFile_2kb.csv'

with open(url, encoding='utf8', errors='ignore') as new_url:
  df_dataset = pd.read_csv(new_url)

df_dataset.head()

#--
def origin(num):
  if num == 1:
    return 'Hello'
  elif num == 2:
    return 'Oi'
  elif num == 3:
    return 'Hola'

# use apply function
df_dataset['3'] = df_dataset['3'].apply(origin)
df_dataset.head()

# using seaborn
sns.pairplot(df_dataset[['3', '35', '38.94']], hue='38.94', height=4)
```
