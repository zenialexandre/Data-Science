```python
import matplotlib.pyplot as plt
import seaborn as sns

# data
cause = 'Chronic Disease', 'Disease2', 'Disease3', 'Disease4'
percentile = [62, 5, 4, 2]

# setting pie chart
plt.figure(figsize=(10, 10))
explode = (0.05, 0, 0, 0)
plt.pie(percentile, labels=cause, explode=explode, autopct='%1.1f%%', startangle=70)
plt.axis('equal')
plt.title('Causes of death of somewhere')
```
