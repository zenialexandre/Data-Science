```python
import matplotlib.pyplot as plt
%matplotlib inline
# to view the plot on notebook

cause = ['Disease1', 'Disease2', 'Disease3', 'Disease4', 'Disease5', 'Disease6']
percentile = [62, 5, 4, 2, 1, 26]

# size of the figure
plt.figure(figsize=(10, 10))

# explode the largest value
explode = (0.05, 0, 0, 0, 0, 0)

# pie properties
plt.pie(percentile, labels=cause, explode=explode, autopct='%1.1f%%', startangle=70)

# setting axis to make a circle
plt.axis('equal')
plt.title('Ohio State - 2012 : Leading Causes of Death')
plt.show()
```
