```python
import matplotlib.pyplot as plt
from matplotlib import style

# define temperature, wind, humidty, precipitation data and time hrs data
temp_data = [79, 75, 74, 72, 81]
wind_data = [14, 12, 10, 13, 9]
time_hrs = [1, 2, 3, 4, 5]
humidity_data = [73, 74, 81, 82, 72]
precipitation_data = [26, 42, 33, 44, 32]

# draw subplots for (1, 2, 1) and (1, 2, 2)
plt.figure(figsize=(8, 4)) # width & height
plt.subplots_adjust(hspace=.25)
plt.subplot(1, 2, 1)
plt.title('Temperature')
plt.plot(time_hrs, temp_data, color='b', linestyle='--', linewidth=1)
#--
plt.subplot(1, 2, 2)
plt.title('Wind')
plt.plot(time_hrs, wind_data, color='r', linestyle='--', linewidth=1)

# subplots for (2, 1, 1) and (2, 1, 2)
plt.figure(figsize=(6, 6))
plt.subplots_adjust(hspace=.25)
plt.subplot(2, 1, 1)
plt.title('Humidity')
plt.plot(time_hrs, humidity_data, color='g', linestyle='-', linewidth=1)
plt.subplot(2, 1, 2) # this here made the graphs being individuals
plt.title('Precipitation')
plt.plot(time_hrs, precipitation_data, color='r', linestyle='-', linewidth=1)
plt.show()

# subplots for (2, 2, 1), (2, 2, 2), (2, 2, 3), (2, 2, 4)
plt.figure(figsize=(9, 9))
plt.subplots_adjust(hspace=.3)
plt.subplot(2, 2, 1)
plt.title('Temperature (F)')
plt.plot(time_hrs, temp_data, color='b', linestyle='-', linewidth=1)
plt.subplot(2, 2, 2)
plt.title('Wind (MPH)')
plt.plot(time_hrs, wind_data, color='r', linestyle='-', linewidth=1)
plt.subplot(2, 2, 3)
plt.title('Humidity (%)')
plt.plot(time_hrs, humidity_data, color='y', linestyle='-', linewidth=1)
plt.subplot(2, 2, 4)
plt.title('Precipitation (%)')
plt.plot(time_hrs, precipitation_data, color='g', linestyle='-', linewidth=1)
```
